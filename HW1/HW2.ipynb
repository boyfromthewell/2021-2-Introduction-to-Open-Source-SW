```python
class Menu():
    def __init__(self):
        self.orderList={}
        self.total=0
        
        self.index=[]
        self.items=[]
        self.price=[]
        self.num=[]
        
        self.orderResult=[[0],[0]]
        self.manageResult=[[0],[0]]
        
    def addMenu(self):
        i=0
        f=open("cafe.txt", "r", encoding="UTF-8")
        for line in f.readlines():
            line=line.strip()
            parts=line.split(", ")
            self.orderList[i]=parts
            
            self.index.append(int(parts[0]))
            self.items.append(parts[1])
            self.price.append(int(parts[2]))
            self.num.append(int(parts[3]))
            
            i+=1
        f.close()
        
    def printMenu(self):
        print("=====================================")
        for i in self.orderList:
            print("{}.{}   금액: {}원 재고: {}".format(self.index[i], self.items[i], self.price[i], self.num[i]))
        print("end를 입력하시면 주문서로 돌아갑니다.")
        print("====================================")
        
    def calOrder(self, menuNum, n):
        self.num[menuNum-1]-=n
        self.orderResult[0][0]+=n
        self.orderResult[1][0]+=n*self.price[menuNum-1]
        self.printMenu()

    def calManage(self, menuNum, n):
        self.num[menuNum-1]+=n
        self.manageResult[0][0]+=n
        self.manageResult[1][0]+=n*self.price[menuNum-1]
        self.printMenu()
    
class Order(Menu):
    
    ###############메뉴 주문관리###############
    
    def printMenu(self):
        print("=====================================")
        for i in self.orderList:
            print("{}.{}   금액: {}원 재고: {}".format(self.index[i], self.items[i], self.price[i], self.num[i]))
        print("end를 입력하시면 주문서로 돌아갑니다.")
        print("====================================")
        
    def orderMenu(self, menuNum):
        if menuNum not in self.index:
            print("존재하지 않는 메뉴 입니다")
        else:
            n=int(input("선택한 메뉴의 수량을 입력해 주세요 : "))
            
            if n>self.num[menuNum-1]:
                if self.num[menuNum-1]==0:
                    print("품절되었습니다 다른 메뉴를 선택해주세요.\n")
                else:
                    print("수량이 부족합니다.\n")
            else:
                self.calOrder(menuNum, n)
            
            while True:
                additionalMenu=input("추가로 주문할 메뉴 번호를 입력해주세요 : ")
                if additionalMenu=="end":
                    print("\n주문내역")
                    print("총 주문 수량: {} 개, 총 주문 금액: {}원".format(self.orderResult[0][0], self.orderResult[1][0]))
                    break;
                additionalMenu=int(additionalMenu)
                if additionalMenu not in self.index:
                    print("존재하지 않는 메뉴 입니다\n")
                else:
                    addN=int(input("선택한 메뉴의 수량을 입력해 주세요 : "))
                    
                    if addN>self.num[additionalMenu-1]:
                        if self.num[additionalMenu-1]==0:
                            print("품절되었습니다 다른 메뉴를 선택해주세요.\n")
                        else:
                            print("수량이 부족합니다.\n")
                    else:
                        self.calOrder(additionalMenu, addN)
                        self.total=self.orderResult[1][0]
                        
    #############################################
    
    ##################입고 관리####################
        
    def management(self, menuNum):
        if menuNum not in self.index:
            print("존재하지 않는 메뉴 입니다")
        else:
            n=int(input("선택한 메뉴의 수량을 입력해 주세요 : "))
            self.calManage(menuNum, n)
            
            while True:
                additionalMenu=input("추가로 주문할 메뉴 번호를 입력해주세요 : ")
                if additionalMenu=="end":
                    print("\n주문내역")
                    print("총 주문 수량: {} 개, 총 주문 금액: {}원".format(self.manageResult[0][0], self.manageResult[1][0]))
                    break;
                additionalMenu=int(additionalMenu)
                if additionalMenu not in self.index:
                    print("존재하지 않는 메뉴 입니다\n")
                else:
                    addN=int(input("선택한 메뉴의 수량을 입력해 주세요 : "))
                    self.calManage(additionalMenu, addN)
                    
    def toSale(self, total):
        print("총 매출은 {}원입니다.\n".format(total))
        
    ####################################################
                        
O=Order()

O.addMenu()

while True:
    print("=================")
    print("     주문서      ")
    print("=================")
    print("1. 커피 주문하기")
    print("2. 커피 매출 확인")
    print("3. 커피 입고 하기")
    print("4. 종료하기")
    print("==================")

    n=input("원하시는 주문 번호를 입력해주세요 : ")
    
    if n=="1":
        O.printMenu()
        menu=int(input("주문할 메뉴 번호를 입력해 주세요 : "))
        O.orderMenu(menu)
    elif n=="2":
        O.toSale(O.total)
    elif n=="3":
        O.printMenu()
        menu=int(input("입고할 메뉴 번호를 입력해 주세요 : "))
        O.management(menu)
    elif n=="4":
        print("종료합니다.")
        break
    else:
        print("잘못된 입력입니다 다시 입력하세요.")
        
```
