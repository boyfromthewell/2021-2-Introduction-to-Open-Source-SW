```python
'''
EX1) Make a factorial calculator (n!) function
- The function name: def Fact(num)
- Return value: Factorial result (num > 0), ‘Error’ (num <= 0)
'''

def fac(n):
    if(n<=0):
        return 'error'
    elif(n>1):
        return n*fac(n-1)
    else:
        return 1
a=int(input('num = '))
fac(a)
```

```python
'''
EX2) Create a student.txt file through a function that generates students' grades
- A student function
» Parameters: student name, midterm exam score, final exam score, and assignment score
» Grade: A (90 <= Average score <=100), B (80 <= Average score <90), C (else)
» No return value but creates a student.txt file to append student information
'''

def student(name, exam1, exam2, homework):
    
    ave=(exam1+exam2+homework)/3
    grade=""
    if(ave>=90):
        grade='A'
    elif(ave<90 and ave >=80):
        grade="B"
    else:
        grade="C"
        
    f=open("student.txt", "a")
    data="%s|%d|%d|%d|%d|%s\n"%(name, exam1, exam2, homework, ave, grade)
    f.write(data)
    f.close()
    
f=open("student.txt", "w")
student('Jerry',90,89,99)
student('Eric',56,45,80)
student('Sun',20,79,60)
student('Jacob',40,39,98)
```

```python
'''
EX3) Create a function that opens a students.txt file and allows students to search their
average and grade by their names
- A search function
» Parameters: student name
» No return value but prints the average and grade of a student

'''

def search(name):
    f=open("student.txt" ,"r")
    lines=f.readlines()

    for line in lines:
        item=line.split('|')
        if(item[0]==name):
            print(f"average: {item[4]}, grade:{item[5]}")
        
search('Sun')
search('Jacob')
```

```python
'''
EX1) Create a class “Queue” that stores integers based on the list. There are 5 methods,
such as enqueue, dequeue, peek, isEmpty, and queueSize in a Queue class.
- isEmpty method checks if queue is empty
- queueSize method returns the size of the queue

'''

from random import *

class Queue():
    def __init__(self):
        self.queue=[]
        
    def enqueue(self, data):
        self.queue.append(data)
    def dequeue(self):
        dequeue_object=None
        if self.isEmpty():
            print("Empty")
        else:
            dequeue_object=self.queue[0]
            self.queue=self.queue[1:]
    def peek(self):
        peek_object=None
        if self.isEmpty():
            print("Empty")
        else:
            peek_object=self.queue[0]
        return peek_object
    
    def isEmpty(self):
        is_empty=False
        if len(self.queue)==0:
            is_empty=True
        return is_empty
    
    def queueSize(self):
        if self.isEmpty():
            print("Empty")
        else:
            return len(self.queue)
    
i=0
Q=Queue()

while i<10:
    num = randint(1,2)
    
    print("---%d step---" % (i+1))
    
    if num==1:
        Q.enqueue(i)
    else:
        num=Q.dequeue()
        
    print("Current queue: ", Q.queue)
    print("Peek data: ", Q.peek())
    print("Queue size: ", Q.queueSize())
    
    i+=1
```
